## Dataset Description
================================= <br>
**Header**	**Definition** <br>
1. gameorder	Play order of game in NBA history
2. game_id	Unique ID for each game
3. lg_id	Which league the game was played in
4. _iscopy	Each row of data is tied to a single team for a single game, so _iscopy flags if this game_id has already occured for the opposing team in the same matchup
5. year_id	Season id, named based on year in which the season ended
6. date_game	Game date
7. is_playoffs	Flag for playoff games
8. team_id	Three letter code for team name, from Basketball Reference
9. fran_id	Franchise id. 
10. Multiple team_ids can fall under the same fran_id due to name changes or moves. Interactive is grouped by fran_id.
11. pts	Points scored by team
12. elo_i	Team elo entering the game
13. elo_n	Team elo following the game
14. win_equiv	Equivalent number of wins in a 82-game season for a team of elo_n quality
15. opp_id	Team id of opponent
16. opp_fran	Franchise id of opponent
17. opp_pts	Points scored by opponent
18. opp_elo_i	Opponent elo entering the game
19. opp_elo_n	Opponent elo following the game
20. game_location	Home (H), away (A), or neutral (N)
21. game_result	Win or loss for team in the team_id column
22. forecast	Elo-based chances of winning for the team in the team_id column, based on elo ratings and game location
23. notes	Additional information

## Invalid Values

### Inconsistent Values
Kadang-kadang nilai suatu kolom adalah realistis, namun tidak sesuai dengan nilai di kolom lainnya. <br>
Sebagai contoh, dalam data NBA, kolom "pts", "opp_pts" dan "game_result" harus konsisten. Semisal "game_result" hasilnya "W" (winning) berarti harus "pts" > "opp_pts" (HARUS KONSISTEN)!

```python
# True means Consistent, vice versa
nba[
    (nba['pts']>nba['opp_pts']) &
    (nba['game_result']!='W')
].empty
```

```python
nba[
    (nba['pts']<nba['opp_pts']) &
    (nba['game_result']!='L')
].empty
```

## Combining Multiple Datasets
Menggabungkan data menjadi satu kumpulan data yang siap dianalisis

```python
city_revenues = pd.Series(
    [100,200,300],
    index=['Amsterdam', 'Toronto', 'Tokyo']
)
city_employee_count = pd.Series(
    {
        'Amsterdam':5,
        'Tokyo':8
    }
)
city_data = pd.DataFrame(
    {
        'revenues':city_revenues,
        'employee_count':city_employee_count
    },
    index=['Amsterdam','Toronto','Tokyo']
)
city_data
```

```python
another_city_data = pd.DataFrame(
    {
        'revenues':[900,800],
        'employee_count':[7,2]
    },
    index=['New York','Barcelona']
)
another_city_data
```

```python
# concat
all_city_data = pd.concat([city_data,another_city_data],sort=False)
all_city_data
```

```python
# Menambahkan kolom dengan axis=1
city_countries = pd.DataFrame(
    {
        'country':['Holland','Japan','Holland','Canada','Spain'],
        'capital' : [1,1,0,0,0]
    },
    index=['Amsterdam', "Tokyo",'Rotterdam', "Toronto", "Barcelona"]
)
cities = pd.concat([all_city_data,city_countries],axis=1,sort=False)
```

```python
cities
```

```python
# Buat dataframe baru yang menunjukkan populasi dan benua masing-masing kota
countries = pd.DataFrame(
    {
        'population_millions':[17,127,37],
        'contingent':['Europe','Asia','North America']
    },
    index = ['Holland','Japan','Canada']
)
# using merge like SQL join='country' (menunjukkan kolom mana yang ingin kalian join) --> default inner join
pd.merge(cities,countries,left_on="country",right_index=True)
```

```python
# using merge like SQL join='country' (menunjukkan kolom mana yang ingin kalian join) --> left join
pd.merge(cities,countries,left_on="country",how="left",right_index=True)
```
