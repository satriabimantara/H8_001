## Plotting in Pandas (Area Plots)
***
Area plot digunakan untuk memvisualisasikan plot sebagai cumulative plot. <br>
Secara default, area plot akan ter-stacked. Semua kolom harus berupa semua nilai positif atau semua nilai negatif. <br>
Nilai NaN akan otomatis ditetapkan secara default ke 0. <br>
Untuk menghasilkan unstacked bisa set unstacked=False

```python
# map index to int for plotting purpose
df_top_5.index = df_top_5.index.map(int)
# Before unstacked
df_top_5.plot(
    kind='area',
    stacked=True,
    figsize=(20,8)
)
plt.title('Trends of top 5 countries that Contributed the most to immigration to Canada')
plt.xlabel('Years')
plt.ylabel('Number of immigrant')

# After unstacked
df_top_5.plot(
    kind='area',
    stacked=False,
    alpha=0.25, #tranparent 0-1 default a=0.5
    figsize=(20,8)
)
plt.title('Trends of top 5 countries that Contributed the most to immigration to Canada')
plt.xlabel('Years')
plt.ylabel('Number of immigrant')
```

## Two Types of Plotting using Matplotlib
***
1. Option 1 (Scripting Layer) <br>
Option ini menggunakan matplotlib.pyplot as 'plt'. Dengan menggunakannya, kita dapat menambahkan lebih banyak elemen dengan memanggil metode yang berbeda secara prosedural: <br>
misalnya plt.title(), plt.xlabel(), dan lainnya <br>
2. Option 2 (Artist Layer) <br>
Option ini disebut sebagai metode object oriented. Menggunakan instansiasi Axed dari Matplotlib. <br>
Instansiasi Axes tersebut ditampung ke dalam sebuah variabel, misalnya 'ax'. Terjadi perubahan kode untuk mengubah title, xlabel, ylabel. <br>
ax.set_title(), ax.set_xlabel(), dan lainnya. <br>
Perubahan terjadi dengan menambahkan "set_" ke metode pada Option 1 <br>
Opsi ini disarankan untuk digunakan karena lebih transparan dan fleksibel untuk digunakan pada plot tingkat lanjut

### Option 1

```python
df_top_5.plot(
    kind='area',
    stacked=False,
    alpha=0.25, #tranparent 0-1 default a=0.5
    figsize=(20,8)
)
plt.title('Trends of top 5 countries that Contributed the most to immigration to Canada')
plt.xlabel('Years')
plt.ylabel('Number of immigrant')
```

# Histogram

## Description
***
Histogram adalah cara untuk menggambarkan distribusi frekuensi dari dataset numerik. <br>
Cara kerjanya adalah : <br>
1. Mempartisi sumbu x menjadi beberapa bin <br>
2. Menetapkan setiap titik data dalam kumpulan data kita ke sebuah bin <br>
3. Menghitung jumlah titik data yang telah ditetapkan ke setiap bin <br>

Sumbu y adalah frekuensi atau banyaknya titik data pada setiap bin. <br>
**Note**
Kita bisa mengubah ukuran bin dan biasanya kita perlu menyesuaikan agar distribusinya ditampilkan dengan baik. <br>

## Example
***
What is the frequency distribution of the number (population) of new immigrants from the various countries to Canada in 2013?

```python
import numpy as np
```

```python
df_can = df_canada['2013']
```

```python
# periksa data yang akan dibagi ke dalam beberapa interval menggunakan Numpy
count, bin_edges = np.histogram(df_can)
```

***
Default, metode histogram memecah dataset menjadi 10 bin <br>
How to interprete this output <br>

> etc <br>

### Using Anotate Just For Scripting Layer
***
Terjadi kenaikan pada tahun 2008 - 2013 (dampak krisis keuangan) <br>

**Notes** <br>
Let's annotate this on the plot using the annotate method of the scripting layer or the pyplot interface. We will pass in the following parameters:

1. s: str, the text of annotation.
2. xy: Tuple specifying the (x,y) point to annotate (in this case, end point of arrow).
3. xytext: Tuple specifying the (x,y) point to place the text (in this case, start point of arrow).
4. xycoords: The coordinate system that xy is given in - 'data' uses the coordinate system of the object being annotated (default).
5. arrowprops: Takes a dictionary of properties to draw the arrow:
6. arrowstyle: Specifies the arrow style, '->' is standard arrow.
7. connectionstyle: Specifies the connection type. arc3 is a straight line.
8. color: Specifes color of arror.
9. lw: Specifies the line width. <br>

Let's also annotate a text to go over the arrow. We will pass in the following additional parameters:

1. rotation: rotation angle of text in degrees (counter clockwise)
2. va: vertical alignment of text [â€˜centerâ€™ | â€˜topâ€™ | â€˜bottomâ€™ | â€˜baselineâ€™]
3. ha: horizontal alignment of text [â€˜centerâ€™ | â€˜rightâ€™ | â€˜leftâ€™]

```python
df_iceland.plot(kind='bar', figsize=(10, 6)) 

plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.title('Icelandic Immigrants to Canada from 1980 to 2013')

# Annotate arrow
plt.annotate('',                      # s: str. will leave it blank for no text
             xy=(32, 70),             # place head of the arrow at point (year 2012 , pop 70)
             xytext=(28, 20),         # place base of the arrow at point (year 2008 , pop 20)
             xycoords='data',         # will use the coordinate system of the object being annotated 
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='blue', lw=2)
            )

# Annotate Text
plt.annotate('2008 - 2011 Financial Crisis', # text to display
             xy=(28, 30),                    # start the text at at point (year 2008 , pop 30)
             rotation=72.5,                  # based on trial and error to match the arrow
             va='bottom',                    # want the text to be vertically 'bottom' aligned
             ha='left',                      # want the text to be horizontally 'left' algned.
            )

plt.show()
```

## Definition

kind='pie'

## Exercise 
***
Gunakan pie chart untuk menjelajahi proporsi imigran baru yang dikelompokkan berdasarkan benua untuk seluruh periode waktu dari 1980 - 2013 <br>

**Tips** <br>
Gunakan metode pandas *groupby* untuk meringkas data imigrasi berdasarkan benua "Continent" <br>
Proses umum dalam *groupby*: <br>

1. **Split** : Membagi data berdasarkan group tertentu
2. **Apply** : Menerapkan fungsi matematika seperti, sum, mean, std(), dan lainnya
3. **Combine** : Menggabungkan hasilnya

```python
df_continent = df_canada.groupby('Continent', axis=0).sum()
df_continent['Total'].head()
```

### Plotting using Pie Chart
***
**TIPS!**
we will pass in kind = 'pie' keyword, along with the following additional parameters: <br>

1. autopct - is a string or function used to label the wedges with their numeric value. The label will be placed inside the wedge. If it is a format string, the label will be fmt%pct.
2. startangle - rotates the start of the pie chart by angle degrees counterclockwise from the x-axis.
3. shadow - Draws a shadow beneath the pie (to give a 3D feel).

```python
df_continent['Total'].plot(kind='pie',
                            figsize=(5, 6),
                            autopct='%1.1f%%', # add in percentages
                            startangle=90,     # start angle 90Â° (Africa)
                            shadow=True,       # add shadow      
                            )

plt.title('Immigration to Canada by Continent [1980 - 2013]')
plt.axis('equal') # Sets the pie chart to look like a circle.

plt.show()
```

#### Mempercantik visualisasi
***
The above visual is not very clear, the numbers and text overlap in some instances. Let's make a few modifications to improve the visuals: <br>

1. Remove the text labels on the pie chart by passing in legend and add it as a seperate legend using plt.legend().
2. Push out the percentages to sit just outside the pie chart by passing in pctdistance parameter.
3. Pass in a custom set of colors for continents by passing in colors parameter.
4. Explode the pie chart to emphasize the lowest three continents (Africa, North America, and Latin America and Carribbean) by pasing in explode parameter.

```python
colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
explode_list = [0.1, 0, 0, 0, 0.1, 0.1] # ratio for each continent with which to offset each wedge.

df_continent['Total'].plot(kind='pie',
                            figsize=(15, 6),
                            autopct='%1.1f%%', 
                            startangle=90,    
                            shadow=True,       
                            labels=None,         # turn off labels on pie chart
                            pctdistance=1.12,    # the ratio between the center of each pie slice and the start of the text generated by autopct 
                            colors=colors_list,  # add custom colors
                            explode=explode_list # 'explode' lowest 3 continents
                            )

# scale the title up by 12% to match pctdistance
plt.title('Immigration to Canada by Continent [1980 - 2013]', y=1.12) 

plt.axis('equal') 

# add legend
plt.legend(labels=df_continent.index, loc='upper left') 

plt.show()
```

# Boxplot

## Definition
***
A box plot is a way of statistically representing the distribution of the data through five main dimensions: <br>

1. Minimun: Smallest number in the dataset.
2. First quartile: Middle number between the minimum and the median.
3. Second quartile (Median): Middle number of the (sorted) dataset.
4. Third quartile: Middle number between median and maximum.
5. Maximum: Highest number in the dataset.<br>

Untuk membuat boxplot, kita bisa menggunakan kind='box' pada plot yang dipanggil pada Series pandas atau DataFrame
***


## Case Study
***
Buat boxplot untuk para imigran Japan antara 1980-2013

```python
years = years
countries = ['Japan']
df_japan = df_canada.loc[countries,years].transpose()
```

```python
# Show describe to know percentile
df_japan.describe()
```

```python
# Plotting
df_japan.plot(kind='box', figsize=(8, 6))

plt.title('Box plot of Japanese Immigrants from 1980 - 2013')
plt.ylabel('Number of Immigrants')

plt.show()
```

We can immediately make a few key observations from the plot above:

1. The minimum number of immigrants is around 200 (min), maximum number is around 1300 (max), and median number of immigrants is around 900 (median).
2. 25% of the years for period 1980 - 2013 had an annual immigrant count of ~500 or fewer (First quartile).
3. 75% of the years for period 1980 - 2013 had an annual immigrant count of ~1100 or fewer (Third quartile).

# Subplot

## Definition
***
Kita bisa membuat beberapa plot dalam figure yang sama. Semisal kita ingin membandingkan boxplot berdampingan dengan line plot imigran China dan India. <br>
Untuk memvisualisasikan beberapa plot bersama-sama, kita bisa membuat **figure** (overall canvas) dan membaginya menjadi *subplots*. masing-masing berisi plot. Dengan subplots, biasanya menggunakan **artist layer** ketimbang **scripting layer**
***
Secara umum sintaks yang digunakan adalah: <br>
fig = plt.figure() # create figure <br>
ax = fig.add_subplot(nrows, ncols, plot_number) # create subplots <br>

Dimana, <br>
1. nrows and ncols are used to notionally split the figure into (nrows * ncols) sub-axes,
2. plot_number is used to identify the particular subplot that this function is to create within the notional grid.
3. plot_number starts at 1, increments across rows first and has a maximum of nrows * ncols as shown below. <br>



## Example Subplot
***
we will visualize df_china_india (boxplot and lineplot using subplot)

```python
n_rows = 1
n_cols = 2

fig = plt.figure()
ax0 = fig.add_subplot(n_rows,n_cols,1) #add subplot 1 (1 row, 2 columns, first plot)
ax1 = fig.add_subplot(n_rows,n_cols,2) #add subplot 1 (1 row, 2 columns, Second plot)

# Subplot 1 (Boxplot)
df_china_india.plot(
    ax=ax0,
    kind='box',
    figsize=(20,6),
    color='blue'
)
ax0.set_title('Box Plots of Immigrants from China and India (1980 - 2013)')
ax0.set_xlabel('Number of Immigrants')
ax0.set_ylabel('Countries')

# Subplot 2 (Line)
df_china_india.plot(
    ax=ax1,
    kind='line',
    figsize=(20,6),
)
ax1.set_title ('Line Plots of Immigrants from China and India (1980 - 2013)')
ax1.set_ylabel('Number of Immigrants')
ax1.set_xlabel('Years')
```

# Bubble Plots

## Definition
***
Bubble plot adalah variasi dari scatter plot yang menampilkan tiga dimensi data (x,y,z). Titik data diganti dengan bubbles. Ukuran bubbles akan ditentukan dari variabel 'z' yang dikenal dengan bobot. <br>
Di matplotlib kita bisa menambahkan array atau skalar ke kata kunci *s* pada *plot()*

## Case Study
***
Argentina suffered a great depression from 1998 - 2002, which caused widespread unemployment, riots, the fall of the government, and a default on the country's foreign debt. In terms of income, over 50% of Argentines were poor, and seven out of ten Argentine children were poor at the depth of the crisis in 2002. <br>
Let's analyze the effect of this crisis, and compare Argentina's immigration to that of it's neighbour Brazil. Let's do that using a bubble plot of immigration from Brazil and Argentina for the years 1980 - 2013. We will set the weights for the bubble as the normalized value of the population for each year.

### Retrieve Data

```python
df_bubble = df_canada.loc[['Argentina','Brazil'],years].transpose()
# reset index year agar menjadi column
df_bubble.reset_index(inplace=True)
df_bubble.columns = ['years','Argentina','Brazil']
```

### Normalize Data Argentina and Brazil using Normalization Formula for weight point

```python
norm_brazil = (df_bubble['Brazil'] - df_bubble['Brazil'].min())/(df_bubble['Brazil'].max() - df_bubble['Brazil'].min())
norm_argentina = (df_bubble['Argentina'] - df_bubble['Argentina'].min())/(df_bubble['Argentina'].max() - df_bubble['Argentina'].min())
```

### Plot the data

We will also pass in the weights using the s parameter. Given that the normalized weights are between 0-1, they won't be visible on the plot. Therefore we will:
1. multiply weights by 2000 to scale it up on the graph, and,
2. add 10 to compensate for the min value (which has a 0 weight and therefore scale with x2000).ipynb_checkpoints/

```python
ax0 = df_bubble.plot(
    kind='scatter',
    x='years',
    y='Brazil',
    s=norm_brazil*2000+10,
    color='red',
    alpha=0.5,
    figsize=(20,5)
)
# Menambahkan scatter Argentina ke scatter brazil
ax1 = df_bubble.plot(
    kind='scatter',
    x='years',
    y='Argentina',
    s=norm_brazil*2000+10,
    color='blue',
    alpha=0.5,
    ax=ax0,
    figsize=(20,5)
)

ax0.set_ylabel('Number of Immigrants')
ax0.set_title('Immigration from Brazil and Argentina from 1980 - 2013')
ax0.legend(['Brazil', 'Argentina'], loc='upper left', fontsize='x-large')
```

The size of the bubble corresponds to the magnitude of immigrating population for that year, compared to the 1980 - 2013 data. The larger the bubble, the more immigrants in that year.

From the plot above, we can see a corresponding increase in immigration from Argentina during the 1998 - 2002 great depression. We can also observe a similar spike around 1985 to 1993. In fact, Argentina had suffered a great depression from 1974 - 1990, just before the onset of 1998 - 2002 great depression.

On a similar note, Brazil suffered the Samba Effect where the Brazilian real (currency) dropped nearly 35% in 1999. There was a fear of a South American financial crisis as many South American countries were heavily dependent on industrial exports from Brazil. The Brazilian government subsequently adopted an austerity program, and the economy slowly recovered over the years, culminating in a surge in 2010. The immigration data reflect these events.
