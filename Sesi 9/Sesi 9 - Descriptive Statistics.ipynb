# Introduction

## Pengertian Statistik Deskriptif
***
Statistik deskriptif adalah kegiatan untuk meringkas dan mengatur data sehingga dapat dengan mudah dipahami. Statistik deskriptif membantu kita memahami data, menarik kesimpulan dari data dimana ini adalah langkah awal yang diperlukan dalam mengembangkan Machine Learning khususnya membuat prediksi. 

## Tipe statistik deskriptif
***
Secara garis besar, statistik deskriptif dibagi menjadi dua bagian, yaitu **Pengukuran tendensi sentral** dan **Pengukuran variabilitas (penyebaran)**. Ada beberapa terminologi yang harus dipahami, yaitu: <br>
1. Populasi dan Sample <br>
> *Populasi* adalah sekumpulan dari semua elemen. Populasi ini biasanya sangat banyak, sehingga para statistikawan biasanya menggunakan subkumpulan dari populasi untuk membuat beberapa kesimpulan tentang suatu populasi. <br>
> *Sample* adalah subset dari populasi. Idealnya, sample harus mempertahankan fitur statistik esensial dari populasi
2. Outliers <br>
> *Outliers* adalah titik data yang berbeda secara signifikan dari sebagian besar data yang diambil dair sampel atau populasi. Ada beberapa hal yang menyebabkan data pencilan, seperti: natural variation in data, change in the behavior of the observed system, and error in data collection

## Pengukuran Tendensi Sentral
***
Pengukuran tendensi sentral berupaya untuk menemukan satu bilangan yang paling baik untuk meringkas seluruh rangkaian pengukuran, dimana bilangan ini adalah *central* dari himpunan tersebut. Ada beberapa pengukuran yang bisa digunakan untuk pengukuran tendensi sentral, seperti: <br>
1. Mean <br>
> Mean  lebih rentan terpengaruh terhadap outliers
2. Median <br>
> Median jauh lebih sedikit dipengaruhi oleh outliers dan skewed data daripada rata-rata
3. Modus <br>
> Jika pada kumpulan data ada dua nilai yang muncul dalam jumlah yang sama disebut sebagai **bimodal** <br>
> Jika ada tiga nilai yang muncul dalam jumlah yang sama disebut sebagai **trimodal** <br>
> Untuk n modus disebut sebagai **multimodal**

## Pengukuran Penyebaran (Spread)
***
Pengukuran ini mengacu pada variabilitas pada data. Pengukuran ini digunakan untuk mengukur *"Seberapa Bagus"* rata-rata mereprensentasikan kumpulan data. <br>
1. Standar Deviasi atau Simpangan Baku <br>
> Menggambarkan seberapa tersebar data dari mean. STD yang rendah menandakan titik data cenderung mendekati nilai *mean* dan begitupun sebaliknya. <br>
> Ketika kita diminta untuk mengukur STD dari beberapa bagian populasi atau segmen populasi, kita menggunkaan rumus **Sample Standar Deviation**, dan ketika mengukur STD dari keseluruhan populasi, kita menggunakan rumus **Population Standar Deviasi**
2. Mean Deviation atau Mean Absolute Deviation <br>
> MAD adalah rata-rata perbedaan absolut antara setiap nilai dalam satu set nilai dan rata-rata semua nilai dari set itu 
3. Variance <br>
> Variance adalah kuadrat dari standar deviasi 
4. Range atau Rentang <br>
> Salah satu teknik statistik deskriptif yang paling sederhana yang mengukur perbedaan antara nilai terendah dan tertinggi
5. Percentile <br>
> Cara untuk merepresentasikan posisi suatu nilai dalam kumpulan data. Nilai kumpulan data harus diurut dalam keadaan menaik sebelum dihitung menggunakan persentil.
6. Quartile <br>
> Nilai yang membagi data ke dalam beberapa quarters (seperempat bagian) dengan data sudah diurutkan dari kecil ke besar. IQR diukur dengan IQR = Q3-Q1
7. Skewness <br>
> Ukuran asimeteri disbutri probabilitas dari real valued random variable tentang meannya. Nilai kemiringan ini bisa positif, negatif atau undefined. <br>
>  <br>
> **Skewness** dihitung menggunakan dua cara: <br>
> 1. Pearson First Coefficient of Skewness (Mode skewness) 
> 2. Pearson Second Coefficient of Skewness (Median skewness) 
> **Interpretasi Skewness**:
>> 1. Arah kemiringan akan diberi tanda. Nol berarti tidak memiliki kemiringan (distribusi normal), tanda negatif artinya *negative skew* dan tanda positif artinya *positive skew*
>> 2. *Coffiecient* membandingkan distribusi sample dengan distribusi normal. Semakin besar nilainya, maka semakin besar perbedaan distribusi dari distribusi normal
8. Kurtosis <br>
> Kurtosis adalah ukuran apakah data bersifat *heavy-tailed* (banyak outliers) atau *light-tailed* (kurang outliers) relatif terhadap distribusi normal. <br>
> 
> **Skewness** menandakan seberapa simetri dan **Kurtosis** mengacu pada tingkat keberadaan data pencilan dalam distribusi
> Ada tiga tipe kurtosis: <br>
>> 1. Mesokurtic <br>
>> distribusi yang memiliki kurtosis serupa dengan kurtosis berdistribusi normal, yaitu nol <br>
>> 2. Leptokurtic <br>
>> Distribusi yang memiliki kurtosis yang lebih besar dari distribusi Mesokurtik dengan ekor distribusi yang tebal dan berat <br>
>> 3. Platykurtic <br>
>> Distribusi yang memiliki kurtosis yang lebih rendah dari distribsi Mesokurtik.
9. Korelasi <br>

### Weighted Mean
***
Mengalikan setiap titik data dengan bobot yang sesuai, menjumlahkannya dan membagi jumlah yang diperoleh dengan jumlah bobot. Metode ini berguna ketika kita menghitung rata-rata kumpulan data yang berisi item yang muncul dengan frekuensi relatif tertentu. Misalnya dalam himpunan data kita ada 20% data dengan nilai 2, 50% data dengan nilai 4, 30% bernilai 8. Sehingga kita bisa menghitung rata-ratanya: <br>
0.2x2 + 0.5x4 + 0.3x8

```python
x = [8.0, 1, 2.5, 4, 28.0]
w = [0.1, 0.2, 0.3, 0.25, 0.15]

wmean = sum(w[i] * x[i] for i in range(len(x))) / sum(w)
print(wmean)

wmean = sum(x_ * w_ for (x_, w_) in zip(x, w)) / sum(w)
print(wmean)
```

Apabila kita memiliki kumpulan data yang besar, kita bisa menghitung weighted mean dari array NumPy atau Series Pandas.

```python
y, z, w = np.array(x), pd.Series(x), np.array(w)

wmean = np.average(y, weights=w)
print(wmean)

wmean = np.average(z, weights=w)
print(wmean)
```

### Geometric Mean
***
Geometric Mean (GM) adalah akar ke-n dari hasil kali semua n elemen xi dalam dataset 𝑥: ⁿ√(Πᵢ𝑥ᵢ), where 𝑖 = 1, 2, …, 𝑛 <br>
Berapa Geometric Mean dari 2 dan 18? <br>
**Jawab:** <br>
n = 2 --> sehingga akar pangkat dua dari (2 x 18) = akar kuadrat (36) = 6 <br>
GM **berguna** ketika kita ingin membandingkan berbagai hal dengan properti yang sangat berbeda. Sebagai contoh: <br>
Kita ingin membeli kamera di toko online. <br>
1. Kamera A memiliki zoom level 200 dan ada 8 review terbaik
2. Kamera B memiliki zoom level 250 dan ada 6 review
Jika kita menggunakan Aritmetika Mean, maka AM untuk kamera A adalah 104 dan kamera B adalah 128. Hal ini karena zoom memiliki angka yang sangat besar, sehingga rating pengguna diabaikan. <br>
Lain halnya jika kita menggunakan GM, maka hasilnya adalah kamera A memiliki GM 40 dan kamera B 38,7, sehingga meskipun kamera B memiliki zoom level 50 kali lebih besar, namun rating pengguna tetap diprioritaskan dalam perhitungannya.

### Harmonic Mean
***
Secara teknis, pengertian dari harmonic mean adalah *the reciprocal of the average of the reciprocal* <br>
**Reciprocal** adalah 1/value. <br>
Harmonic mean adalah reciprocal dari mean dari semua item dalam dataset: 𝑛 / Σᵢ(1/𝑥ᵢ), where 𝑖 = 1, 2, …, 𝑛 and 𝑛 is the number of items in the dataset 𝑥. <br>
Harmonic mean juga bagus untuk handling outliers yang besar. Semisal ada data {2,4,6,100}, n = 4 <br>
HM = 4/(1/2 + 1/4 + 1/6 + 1/100) = 4.32 <br>
Aritmetika Mean = 2+4+6+100/4 = 28

### Sample Variance
***
*Sample Variance* mengukur penyebaran data. Nilai ini menunjukkan secara numerik **seberapa jauh titik data dari mean**. Secara matematis, rumus *sample variance* adalah: 𝑠² = Σᵢ(𝑥ᵢ − mean(𝑥))² / (𝑛 − 1), where 𝑖 = 1, 2, …, 𝑛 and mean(𝑥) is the sample mean of 𝑥. <br>
Gambar di bawah menunjukkan seberapa penting mengukur variance di dalam mendeskripsikan dataset: <br>
 <br>
Terdapat dua dataset pada gambar diatas, dimana kedua dataset itu memiliki *mean* dan *median* yang sama. Oleh karena itu, kedua pengukuran itu tidaklah cukup. Dengan *variance* kita bisa mengetahui bahwa dataset 1 memiliki variance yang lebih kecil karena titiknya cenderung lebih dekat dengan mean, sebaliknya dataset 2 memiliki variance yang lebih besar karena titiknya cenderung lebih jauh dari mean. Selain itu, range dari dataset 2 juga lebih besar

```python
# Menghitung variance manual
n = len(x)

mean_ = sum(x) / n

var_ = sum((item - mean_)**2 for item in x) / (n - 1)
var_
```

```python
# using statistics 
var = stats.variance(x)
var
```

```python
# using numpy
var_np = np.var(np.array(x), ddof=1)
var_np
```

*ddof* adalah degree of freedom. Nilai 1 menyebabkan penyebut dalam perhitungan variancenya adalah (n-1) dan bukan *n*

## Skewness
***
Kegunaannya mengukur **kesimetrisan** sample data. Ada beberapa definisi matematis tentang skewness.
1. Satu ekspresi umum untuk menghitung skewness kumpulan data x dengan elemen *n* adalah:
> (𝑛² / ((𝑛 − 1)(𝑛 − 2))) (Σᵢ(𝑥ᵢ − mean(𝑥))³ / (𝑛𝑠³))
2. Ekspresi yang lebih sederhana adalah 
> Σᵢ(𝑥ᵢ − mean(𝑥))³ 𝑛 / ((𝑛 − 1)(𝑛 − 2)𝑠³), where 𝑖 = 1, 2, …, 𝑛 and mean(𝑥) is the sample mean of 𝑥

Jika *ekor distribusi dominan di sisi kiri* maka disebut **negative skewness** dan sebaliknya. Jika skewness mendekati 0 (misalnya antara -0.5 dan 0.5), maka dataset dianggap cukup simetris. <br>
Setelah kita menghitung nilai *sample mean*, *standard deviation* dan mengetahui jumlah data *n*, kita bisa mendapatkan sample skewness dengan Python:

```python
# gunakan rumus ekspresi yang lebih sederhana
x = [8.0, 1, 2.5, 4, 28.0]

n = len(x)

mean_ = sum(x) / n
var_ = sum((item - mean_)**2 for item in x) / (n-1)
std_ = var_ ** 0.5

skew_ = (sum((item - mean_)**3 for item in x) * n / ((n - 1) * (n - 2) * std_**3))
```

```python
skew_ #this right tail
```

```python
# menghitung skew menggunakan scipy stats
scistats.skew(y,bias=False)
```

Parameter *bias* di set ke **False** untuk mengaktifkan koreksi bias statistik. Ketika menghitung skew untuk array yang mengandung *nan* maka secara default akan mengembalikan *nan*

```python
# Pandas Series juga memiliki metode .skew()
z.skew()
```

## Mengukur korelasi antara pasangan data
***
Memeriksa hubungan antara elemen yang sesuai dari dua variabel dalam kumpulan data. Semisal ada dua variabel X dan Y dan di variabel X ada elemen (x1,x2,...,xn) dan di variabel Y ada elemen (y1,y2,...,yn). Maka ada sejumlah n pasang elemen yang bersesuaian (x1,y1), (x2,y2),...,(xn,yn). <br>
Ada tiga jenis pengukuran korelasi: <br>
1. Positive correlation
2. Negative correlation
3. Weak or no correlation

Dua statistik untuk mengukur korelasi dataset adalah **covariance** dan **correlation coefficient**

```python
# inisialisasi data 
x = list(range(-10, 11))
y = [0, 2, 2, 2, 2, 3, 3, 6, 7, 4, 7, 6, 6, 9, 4, 5, 5, 10, 11, 12, 14]
x_, y_ = np.array(x), np.array(y)
x__, y__ = pd.Series(x_), pd.Series(y_)
```

### Covariance
***
*Sample Covariance* adalah ukuran yang mengukur kekuatan dan arah hubungan antara sepasang variabel. <br>
1. Jika **korelasi positif**, maka **covariance positif**. Hubungan yang kuat menandakan nilai yang tinggi dari covariance
2. Jika **korelasi negatif**, maka **covariance negatif**.
3. Jika **korelasi lemah (close to zero)**, maka **covariance mendekati 0**

```python
# hitung covariance matrix dengan NumPy
cov_matrix = np.cov(x_,y_)
cov_matrix
```

Elemen cov_matrix(0,0) menandakan covariance dari x dengan x, begitupun cov_matrix(1,1) menandakan covariance dari y dengan y. Elemen cov_matrix(0,1) dan cov_matrix(1,0) adalah covariance aktual antara x dan y

```python
x_.var(ddof=1)
```

```python
y_.var(ddof=1)
```

Series Pandas memiliki metode .cov() yang dapat digunakan untuk menghitung kovarian

```python
cov_xy = x__.cov(y__)
cov_xy
```

### Correlation Coefficient
***
Disebut juga sebagai *Pearson product-moment correlation* yang dilambangkan dengan simbol *r*. 
1. r>0 mengindikasikan korelasi positif
2. r<0 mengindikasikan korelasi negatif
3. r = 1 nilai maksimum yang mungkin dari r (perfect positive linear relationship between variables)
4. r = -1 nilai minimum yang mungkin dari r (perfect negative linear relationship between variables)
5. r mendekati 0 atau 0 artinya korelasi diantara variable lemah

```python
# calculate correlation coefficient dengan NumPy
corr_matrix = np.corrcoef(x_,y_)
corr_matrix
```

```python
# Series Pandas memiliki method .corr()
r = x__.corr(y__)
r
```

3. **showmeans** show the means of your data when True
4. **meanline** represent the mean as line when True
5. **labels** the labels of your data
6. **medianprops** denotes the properties of the line representing the median
